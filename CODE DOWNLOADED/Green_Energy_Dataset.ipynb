```python
import pandas as pd
import numpy as np
```

```python
# Loading the dataset
df = pd.read_csv('/content/green_energy_dataset.csv')
```

```python
# Standardising the column names
# Removing the spaces
# Replacing spaces with underscores
# Replacing % with 'Percent'
# Replacing slashe
# Cleaning up double underscores
df.columns = (
    df.columns.str.strip()
             .str.replace(" ", "_")
             .str.replace("%", "Percent")
             .str.replace("/", "_")
             .str.replace("__", "_"))
```

```python
# Inspecting and handling the missing values
missing_counts = df.isnull().sum()
```

```python
# Checking the data types
print(df.dtypes)
```

```python
# Removing the illogical values
df = df[(df['Year_Built'] >= 1700) & (df['Floor_Area_m2'] > 0)]
```

```python
# Energy use per square meter
df['Energy_Demand_per_m2'] = df['Energy_Demand_kWh'] / df['Floor_Area_m2']
```

```python
# Solar energy potential
df['Solar_Energy_Capacity_kWh'] = df['Solar_Potential_kWh_m2'] * df['Installation_Area_m2']
```

```python
# Carbon savings in actual energy
df['Carbon_Reduction_kWh'] = df['Energy_Demand_kWh'] * (df['Carbon_Reduction_Percent'] / 100)
```

```python
# Categorizing the continuous data for filtering
df['Building_Size_Category'] = pd.cut(df['Floor_Area_m2'],
                                      bins=[0, 100, 200, 300],
                                      labels=['Small', 'Medium', 'Large'])
```

```python
# For Saving cleaned dataset
df.to_csv("cleaned_green_energy_dataset.csv", index=False)
```
